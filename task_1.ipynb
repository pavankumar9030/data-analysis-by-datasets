```python
import pandas as pd
#import packages
df = pd.read_csv("https://raw.githubusercontent.com/arunk13/MSDA-Assignments/master/IS607Fall2015/Assignment3/student-por.csv", sep=';')
#import csv file i import this file from the github because the code didnt run untile we enter the csv file that are existed in our drive
print("First 5 rows of the dataset: ")
#columns
print(df.head())
```

```python
print("missing values in colume")
print(df.isnull().sum())
#misiing values
print("\n")
print("colume data types")
print(df.dtypes)
#data types
print("\n")
print("understand the dataset:")
print(df.shape)
#shape(rows,column)
```

```python
before = df.shape[0]
#replace before the rows

df = df.drop_duplicates()
#for the data cleaning

after = df.shape[0]
#replace after the rows

print(f"Removed {before - after} entire removed duplicates.")
print("Final dataset:", df.shape)
print("\n")
#print final dataset
```

```python
import matplotlib.pyplot as plt
#package to draw or visualized for graphs in 2D
plt.figure(figsize=(6,4))
plt.hist(df["G3"], bins=10, color='red', edgecolor='black')
#hist= Histogram
plt.title("Histogram of Final Grades (G3)")
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.show()
#print histogram
```

```python
import matplotlib.pyplot as plt
#import package
plt.figure(figsize=(6,4))
plt.scatter(df["studytime"], df["G3"], alpha=0.6, color='black')
#scatter=scatter plot
plt.title("Study Time vs Final Grade (G3)")
#title bar
plt.xlabel("Study Time ")
plt.ylabel("Final Grade (G3)")
plt.show()
```

```python
import matplotlib.pyplot as plt

gender_avg = df.groupby('sex')['G3'].mean()
#import and megre two items sex and G3 together a bar graph

plt.figure(figsize=(6,4))
gender_avg.plot(kind="bar", color=['lightgreen','lightcoral'], edgecolor='black')
#clour and type of graph
plt.title("Average Final Grade (G3) by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Final Grade (G3)")
plt.show()#print bar graph
```
